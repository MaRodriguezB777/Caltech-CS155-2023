```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/emiletimothy/Caltech-CS155-2023/main/set2/data/problem3data.txt', 'problem3data.txt')

train_data = np.loadtxt('problem3data.txt')
y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
# Lasso Regression with Regularization

alphas = np.linspace(0.01, 3, 30)
weights = np.empty((len(alphas), 9))

plt.figure()
for ind, _alpha in enumerate(alphas):
    lasso = Lasso(alpha=_alpha)
    weights[ind] = lasso.fit(X_train, y_train).coef_

plt.plot(alphas, weights)
plt.xlabel("Alpha Value")
plt.ylabel("Weight values")
plt.title("Lasso Regularization Weight Plot")
```

```python
# Ridge Regression with Regularization

alphas = np.linspace(1, 1e4, int(1e4))
weights = np.empty((len(alphas), 9))

plt.figure()
for ind, _alpha in enumerate(alphas):
    ridge = Ridge(alpha=_alpha)
    weights[ind] = ridge.fit(X_train, y_train).coef_

plt.plot(alphas, weights)
plt.xlabel("Alpha Value")
plt.ylabel("Weight values")
plt.title("Ridge Regularization Weight Plot")
```
